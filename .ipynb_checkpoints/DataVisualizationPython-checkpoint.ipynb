### Step-1: Importing Packages

```python
#import all librarys and modules needed
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
from matplotlib import style
# setting style for graphs 
style.use('ggplot')
plt.rcParams['figure.figsize'] = (20, 10)
```

```python
canada = pd.read_excel(r'data/Canada.xlsx', 1, skiprows = range(20), skipfooter = 2)
canada.columns
```

### Step-2 : Importing and Cleaning Data 

```python
#import and clean data, remember path is data/Canada.xlsx
canada = pd.read_excel(r'data/Canada.xlsx', 1, skiprows = range(20), skipfooter = 2)

canada.drop(['AREA', 'REG', 'DEV', 'Type', 'Coverage', 'DevName'], axis = 1, inplace = True)
canada.rename(columns = {'OdName': 'country', 'AreaName':'Continent', 'RegName': 'Region'}, inplace = True)
canada['total'] = canada.sum(axis = 1, numeric_only = True)
canada = canada.set_index('country')
canada.rename(index = {'United Kingdom of Great Britain and Northern Ireland' : 'UK & Ireland'}, inplace = True)
canada.columns = canada.columns.astype(str)
# useful for upcoming visualizations
years = list(map(str, range(1980, 2013)))
canada.tail(20)
```

### Step-3 : Creating Beautiful Visualizations

#### Line Chart

```python
# Single line chart

fig1 = canada.loc['Haiti', years].plot(kind = 'line', color = 'r')
plt.title('Immigration from Haiti to Canada from 1980-2013',color = 'black')
plt.xlabel('Years',color = 'black')
plt.ylabel('Number of Immigrants',color = 'black')
plt.xticks(color = 'black')
plt.yticks(color = 'black')
plt.savefig('linechart_single.png')

plt.show()
```

```python
# Multiple Line chart 

fig2 = plt.plot(canada.loc['India',years], label = 'India')
plt.plot(canada.loc['China',years], label = 'China')
plt.plot(canada.loc['Philippines',years], label = 'Sri Lanka')
plt.legend(loc = 'upper left', fontsize = 12)
plt.xticks(rotation = 90, color = 'black')
plt.yticks(color = 'black')
plt.title('Immigration to Canada from 1980-2013',color = 'black')
plt.xlabel('Year',color = 'black')
plt.ylabel('Number of Immigrants',color = 'black')
plt.savefig('linechart_multiple.png')

plt.show()
```

#### Bar Chart

```python
# Vertical bar chart
# Do not change the style back to ggplot 
# delete the style.use('ggplot') line of code

canada_tot = pd.DataFrame(canada.loc[:,years].sum())
canada_tot.rename(columns = {0:'total'}, inplace = True)

canada_tot.plot(kind = 'bar', legend = False)
plt.title('Total Immigrants to Canada from 1980-2013',color = 'white')
plt.xticks(color = 'violet')
plt.yticks(color = 'violet')
plt.xlabel('Year',color = 'white')
plt.ylabel('Number of Immigrants',color = 'white')
plt.savefig('bar_vertical.png')

plt.show()
```

```python
# Horizontal bar chart
#change style back to ggplt

style.use('ggplot')

canada_top10 = pd.DataFrame(canada.nlargest(10,'total')['total'].sort_values(ascending = True))

canada_top10.plot.barh(legend = False, color = 'y', edgecolor = 'b')
plt.title('Top 10 Immigrant Countries to Canada from 1980-2013',color = 'black')
plt.xlabel('Number of Immigrants',color = 'black')
plt.ylabel('Country',color = 'black')
plt.xticks(color = 'black')
plt.yticks(color = 'black')
plt.savefig('bar_horizontal.png')

plt.show()
```

```python
# Grouped bar chart

year_int10 = list(map(str, (1980,1990,2000,2010, 2013)))
canada_group = pd.DataFrame(canada.loc[['India','China','Philippines','Pakistan'],year_int10].T)

canada_group.plot.bar(edgecolor = 'white')
plt.title('Total Immigrants to Canada from 1980-2013',color = 'black')
plt.xticks(color = 'black')
plt.yticks(color = 'black')
plt.xlabel('Year',color = 'black')
plt.ylabel('Number of Immigrants',color = 'black')
plt.legend(title = 'Country', fontsize = 12,  labelcolor='black')
plt.savefig('bar_grouped.png')

plt.show()
```

```python
# stacked area chart

color = ['yellow','violet']

#make variables that select data we want to compare
top2_list = canada.nlargest(2, 'total').index.tolist()
canada_top2 = pd.DataFrame(canada.loc[top2_list, years].T)

#create a plot and select the details
canada_top2.plot(kind = 'area', stacked = True, color = color)
plt.title('Total Immigrants to Canada from 1980-2013',color = 'skyblue')
plt.legend(fontsize = 12)
plt.xlabel('Year',color = 'w')
plt.ylabel('Number of Immigrants',color = 'w')
plt.xticks(color = 'w')
plt.yticks(color = 'w')
plt.savefig('area_stacked.png')

plt.show()
```

#### Scatter Plot

With the newist version of Seaborn we have to specify x and y.

example: sb.scatterplot(x = 'sepal_length', y = 'sepal_width', data = df_iris)

```python
#scatter plot comparing sepal length to sepal width
style.use('ggplot')
canada_iris = sb.load_dataset('iris')

sb.scatterplot(x = 'sepal_length', y = 'sepal_width', data = canada_iris, s = 200, linewidth = 3, edgecolor = 'Red')
plt.title('Scatter Plot sample', color = 'black', fontsize = 18)
plt.savefig('scatter.png')

plt.show()
```

#### Histogram

With the newist version of Seaborn they have deprecated distplot. Replace distplot with histplot and add kde = True

example:
sb.histplot(df_iris['sepal_length'], color = 'Red', label = 'Sepal Length', kde = True)

```python
#Histogram side by side, with kde
canada_iris = sb.load_dataset('iris')

sb.histplot(canada_iris['sepal_length'], color = 'Red', label = 'Sepal Length', kde = True)
sb.histplot(canada_iris['sepal_width'], color = 'skyblue', label = 'Sepal Width', kde = True)
plt.legend(fontsize = 12, labelcolor = 'black')
plt.xlabel('Sepal Width', color = 'black')
plt.ylabel('Frequency', color = 'black')
plt.xticks(color = 'black')
plt.yticks(color = 'black')
plt.savefig('histogram.png')

plt.show()
```

#### Doughnut Chart

```python
# Doughnut Chart

style.use('ggplot')


top5_list = canada.nlargest(5, 'total').index.tolist()
canada_top5 = pd.DataFrame(canada.loc[top5_list, 'total'].T)
circle = plt.Circle( (0,0), 0.7, color='white')
colors = ['gold', 'green', 'magenta', 'lightskyblue', 'lightgreen']

plt.pie(canada_top5['total'], autopct = '%1.1f%%', shadow = True, explode = [0.1,0,0,0,0], colors = colors, startangle = 90)
fig = plt.gcf()
fig.gca().add_artist(circle)
plt.legend(canada_top5.index, fontsize = 12, loc = 'upper left', labelcolor = 'purple')
plt.title('Top 5 Immigrant Country Distribution', color = 'black', fontsize = 18)
plt.axis('equal')
plt.savefig('doughnut.png')

plt.show()
```

#### Regression Plot

```python
# Strong trend

style.use('cyberpunk')

canada_reg = pd.DataFrame(canada.loc['China',years])
canada_reg.reset_index(inplace = True)
canada_reg.rename(columns = {'index':'year'}, inplace = True)
canada_reg[['year','China']] = canada_reg[['year','China']].astype(int)

sb.regplot(x = 'year', y = 'China', data = canada_reg, color = 'orange', scatter_kws = {'s':150,'alpha':0.4})
plt.title('Total Immigration from China to Canada from 1980-2013', color = 'w')
plt.xlabel('Year', color = 'w')
plt.ylabel('Number of Immigrants', color = 'w')
plt.xticks(color = 'w')
plt.yticks(color = 'w')
plt.savefig('reg_strong.png')

plt.show()
```

Feel free to experiment and add any notes or observations here. <br><br><br><br><br>

```python
# Weak trend

style.use('cyberpunk')

canada_reg = pd.DataFrame(canada.loc[['Denmark','Norway','Sweden'],years].sum())
canada_reg.reset_index(inplace = True)
canada_reg.rename(columns = {'index':'year',0:'total'}, inplace = True)
canada_reg[['year','total']] = canada_reg[['year','total']].astype(int)

sb.regplot(x = 'year', y = 'total', data = canada_reg, color = 'magenta', scatter_kws = {'s':150,'alpha':0.4})
plt.title('Total Immigrants from Scandinavia to Canada from 1980-2013', color = 'w')
plt.xlabel('Year', color = 'w')
plt.ylabel('Number of Immigrants', color = 'w')
plt.xticks(color = 'w')
plt.yticks(color = 'w')
plt.savefig('reg_strong.png')

plt.show()
```

```python
# word cloud
from wordcloud import WordCloud, STOPWORDS
text = open(r'data/Alice.txt', 'r', encoding = 'utf-8').read()
stopwords = set(STOPWORDS)
wordcloud = WordCloud(background_color = 'white', max_words = 200, stopwords = stopwords)

wordcloud.generate(text)
plt.imshow(wordcloud, interpolation = 'bilinear')
plt.axis('off')
plt.savefig('wordcloud.png')

plt.show()
```
